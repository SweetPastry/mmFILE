```
(*没有图片是因为在 jupyter 中图像是用 HTML 渲染的, mm 中不是*)

Psi1[x_, t_] := Sin[2 x - 1 t]
Psi2[x_, t_] := Sin[1 x - 2 t]

Psi[x_, t_] := Psi1[x, t] + Psi2[x, t]

Manipulate[
 Plot[Psi[x, t], {x, -10, 10}, PlotRange -> {-2, 2}, 
  AxesLabel -> {"x", "Amplitude"}, 
  PlotLabel -> "Wave function superposition", 
  PlotStyle -> {Blue, Thick}], {t, 0, 10}
]
```
